```python
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import requests
import re
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the website site: https://www.sportsmediawatch.com/2020-nfl-ratings-page/
url = 'https://escharts.com/top-games?order=peak&year=2021'
browser.visit(url)
```

```python
# scrape the website

html = browser.html
s = soup(html, 'lxml') # s was soup


```

```python
# Extract all the text elements
title = s.title.text
print(title)
```

```python

            
            
            
#             csvRow.append(cell.get_text())
#             writer.writerow(csvRow)
# finally:
#     csvFile.close()


def get_views_list(soup):
    list_text = soup.find_all("span", attrs={"class": ["text-xs",'text-right']})
    len(list_text)
    views_list_2021 = []
    for i in range(len(list_text)):
        if re.search(r'^[0-9\s]+$',list_text[i].text):
            print(list_text[i].text)
            views_list_2021.append(list_text[i].text)
    for i in range(len(views_list_2021)):
        views_list_2021[i] = views_list_2021[i].strip()
        views_list_2021[i] = int(views_list_2021[i].replace(' ',''))
    return views_list_2021[:20]

views_dict = {} # year to list of views
for year in range(2017,2022):
    url = f'https://escharts.com/top-games?order=peak&year={year}'
    browser.visit(url)
    html = browser.html
    s = soup(html, 'lxml')
    views_dict[year] = get_views_list(s)





    
    



```
