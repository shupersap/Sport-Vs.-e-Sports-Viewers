```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
from splinter import Browser
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
def main(url):
    try:
        for page in range(1,3):
            browser.visit(url + "&page=" + str(page))
            html = browser.html
            soup = BeautifulSoup(html, 'html.parser')
            table = soup.select("table")[0]
            columns = table.find("thead").find_all("th")
            column_n= [c.get_text().replace("\n","") for c in columns]
            data=[]
            rows= table.find("tbody").find_all("tr")
            for tr in rows:
                td=tr.find_all("td")
                act_row=[str(tr.get_text()).replace("\n","").replace(' ', '').replace("\t","").replace("HW","").replace("PV","") for tr in td]
                data.append(act_row)
            words = ["one","two","three"]
            dictionary_frames = {page: pd.DataFrame(data,columns=column_n) for page in words}
    except IndexError:
        return dictionary_frames
```
