```python
def main(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    table = soup.select("table")[0]
    columns = table.find("thead").find_all("th")
    column_n= [c.get_text().replace("\n","") for c in columns]
    data=[]
    rows= table.find("tbody").find_all("tr")
    for tr in rows:
        td=tr.find_all("td")
        act_row=[str(tr.get_text()).replace("\n","").replace(' ', '').replace("\t","").replace("HW","").replace("PV","") for tr in td]
        data.append(act_row)
    df= pd.DataFrame(data,columns=column_n)
    
    return df
```

```python
main("https://escharts.com/tournaments/lol?order=peak")
```
